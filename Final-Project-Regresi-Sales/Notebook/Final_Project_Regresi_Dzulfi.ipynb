#Feature Engineer


```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day_of_month'] = df['Date'].dt.day
df['week_of_year'] = df['Date'].dt.isocalendar().week.astype(int)
df['quarter'] = df['Date'].dt.quarter
df['is_weekend'] = df['Date'].dt.dayofweek.isin([5, 6]).astype(int)
df['is_payday_period'] = np.where((
    df['day_of_month'] >= 25) | (df['day_of_month'] == 2), 1, 0)
df['event_harbolnas'] = np.where(df['day_of_month'] == df['month'], 1, 0)
df['customer_ratio'] = (
    df['Customers_id'] / df.groupby('Store')['Customers_id'].transform('mean'))
df['kategori_sales'] = pd.qcut(
    df['Sales'],q=4,
    labels=['Low', 'Medium', 'High', 'Very High'])
df['store_avg_customer'] = (
    df.groupby('Store')['Customers_id']
      .transform('mean'))
df['store_avg_sales'] = (
    df.groupby('Store')['Sales']
      .transform('mean'))
```

#Outlier

```python
import seaborn as sns
import matplotlib.pyplot as plt

#melihat ada berapa outliner di dalam dataset
for col in df.select_dtypes(include='number').columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outlier_count = df[(df[col] < lower) | (df[col] > upper)].shape[0]

    if outlier_count > 0:
        print(f"{col} memiliki {outlier_count} outlier")
```

##Ordinal Encoding

```python
mapping = {
    'Low': 1,
    'Medium': 2,
    'High': 3,
    'Very High': 4
}
X_train['kategori_sales'] = X_train['kategori_sales'].map(mapping)
X_test['kategori_sales'] = X_test['kategori_sales'].map(mapping)
```

##One Hot Encoding

```python
# One-hot encode 'StateHoliday'
X_train = pd.get_dummies(X_train, columns=['StateHoliday'], drop_first=True, dtype=int)
X_test = pd.get_dummies(X_test, columns=['StateHoliday'], drop_first=True, dtype=int)
```

##Feature Alignment

```python
mapping = {
    'Low': 1,
    'Medium': 2,
    'High': 3,
    'Very High': 4
}
X_train['kategori_sales'] = X_train['kategori_sales'].map(mapping)
X_test['kategori_sales'] = X_test['kategori_sales'].map(mapping)

# One-hot encode 'StateHoliday'
X_train = pd.get_dummies(X_train, columns=['StateHoliday'], drop_first=True, dtype=int)
X_test = pd.get_dummies(X_test, columns=['StateHoliday'], drop_first=True, dtype=int)

# Align columns after one-hot encoding
train_cols = X_train.columns
test_cols = X_test.columns

missing_in_test = set(train_cols) - set(test_cols)
for c in missing_in_test:
    X_test[c] = 0

missing_in_train = set(test_cols) - set(train_cols)
for c in missing_in_train:
    X_train[c] = 0

X_test = X_test[train_cols]
```

##Feature Selection

```python
# Define drop_cols as it was deleted in previous steps
drop_cols = [
    'Customers_id',
    'year',
    'Open',
    'kategori_sales',
    'quarter',
    'event_harbolnas',
    'store_avg_sales',
    'customer_ratio'
]
X_train = X_train.drop(columns=drop_cols, errors='ignore')
X_test = X_test.drop(columns=drop_cols, errors='ignore')
```

##Feature Scalling

```python
rs = RobustScaler()
X_train_transformed = rs.fit_transform(X_train)
X_test_transformed = rs.transform(X_test)
```

###XGBoost Regression

```python
xgb = XGBRegressor(n_estimators=100,learning_rate=0.1, max_depth=3, random_state=42, enable_categorical=True)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

##Evaluasi Model

```python
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)

result = []

mask = y_test > 0

models = {
    'Linear Regression': y_pred_lr,
    'Ridge Regression': y_pred_ridge,
    'Lasso Regression': y_pred_lasso,
    #'SVR': y_pred_svr,
    'Adaboost Regression': y_pred_ada,
    'XGBoost Regression': y_pred_xgb
}

for name, y_pred in models.items():
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mape = mean_absolute_percentage_error(y_test[mask],y_pred[mask]) * 100
    r2 = r2_score(y_test,y_pred)
    result.append([name,mae,rmse,mape,r2])

comparison_df = pd.DataFrame(
    result,
    columns=['Model','MAE','RMSE','MAPE (%)','R²'])

comparison_df = comparison_df.round(2)
comparison_df
```
